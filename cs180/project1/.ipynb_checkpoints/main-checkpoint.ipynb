```python
def ssd(img1, img2):
    return np.sum((img1 - img2) ** 2)
```

```python
def ncc(img1, img2):
    norm1 = img1 - np.mean(img1)
    norm2 = img2 - np.mean(img2)
    return np.sum(norm1 * norm2) / np.sqrt(np.sum(norm1 ** 2) * np.sum(norm2 ** 2))
```

```python
def single_ssd_offset(b, g, r, window):
    min_l2_bg = float('inf')
    min_l2_br = float('inf')
    best_shift_bg = (0, 0)
    best_shift_br = (0, 0)
    for i in range(window[0], window[1]):
        for j in range(window[0], window[1]):
            shift_g = np.roll(g, (i, j), (0, 1))
            l2_bg = ssd(b, shift_g)
            if l2_bg < min_l2_bg:
                min_l2_bg = l2_bg
                best_shift_bg = (i, j)

            shift_r = np.roll(r, (i, j), (0, 1))
            l2_br = ssd(b, shift_r)
            if l2_br < min_l2_br:
                min_l2_br = l2_br
                best_shift_br = (i, j)

    return best_shift_bg, best_shift_br

def shift_with_offset_and_show(offset, b, g, r, show=True):
    ag = np.roll(g, offset[0], (0, 1))   
    ar = np.roll(r, offset[1], (0, 1))
    im_out = np.dstack([ar, ag, b])
    if show:
        skio.imshow(im_out)
        plt.axis('off')
        plt.show()
    return im_out
```

```python
offset = single_ssd_offset(b, g, r, (-15, 16))
print(offset) # g vs. b and r vs. b
im_out = shift_with_offset_and_show(offset, b, g, r)
```

```python
def single_ncc_offset(b, g, r, window):
    max_l2_bg = -1
    max_l2_br = -1
    best_shift_bg = (0, 0)
    best_shift_br = (0, 0)
    for i in range(window[0], window[1]):
        for j in range(window[0], window[1]):
            shift_g = np.roll(g, (i, j), (0, 1))
            l2_bg = ncc(b, shift_g)
            if l2_bg > max_l2_bg:
                max_l2_bg = l2_bg
                best_shift_bg = (i, j)

            shift_r = np.roll(r, (i, j), (0, 1))
            l2_br = ncc(b, shift_r)
            if l2_br > max_l2_br:
                max_l2_br = l2_br
                best_shift_br = (i, j)

    return best_shift_bg, best_shift_br
```

```python
offset = single_ncc_offset(b, g, r, (-15, 16))
print(offset) # g vs. b and r vs. b
im_out = shift_with_offset_and_show(offset, b, g, r)
```

```python
def crop_rgb(rate, b, g, r):
    crop_row = int(b.shape[0] * rate)
    crop_col = int(b.shape[1] * rate)
    crop_b = b[crop_row : b.shape[0]-crop_row, crop_col : b.shape[1]-crop_col]
    crop_g = g[crop_row : b.shape[0]-crop_row, crop_col : b.shape[1]-crop_col]
    crop_r = r[crop_row : b.shape[0]-crop_row, crop_col : b.shape[1]-crop_col]
    return crop_b, crop_g, crop_r

def crop_img(rate, img):
    crop_row = int(img.shape[0] * rate)
    crop_col = int(img.shape[1] * rate)
    crop_img = img[crop_row : img.shape[0]-crop_row, crop_col : img.shape[1]-crop_col, :]
    return crop_img
```

```python
name_list = ["cathedral.jpg", "monastery.jpg", "tobolsk.jpg"]
imgs_ssd = []
imgs_ncc = []
offsets_ssd = []
offsets_ncc = []

for name in name_list:
    imname = 'data/' + name
    im = skio.imread(imname)
    im = sk.img_as_float(im)
    height = np.floor(im.shape[0] / 3.0).astype(int)
    b = im[:height]
    g = im[height: 2*height]
    r = im[2*height: 3*height]
    crop_b, crop_g, crop_r = crop_rgb(0.1, b, g, r)
    
    offset = single_ssd_offset(crop_b, crop_g, crop_r, (-15, 16))
    im_out = shift_with_offset_and_show(offset, b, g, r, show=False)
    imgs_ssd.append(im_out)
    offsets_ssd.append(offset)
    
    offset = single_ncc_offset(crop_b, crop_g, crop_r, (-15, 16))
    im_out = shift_with_offset_and_show(offset, b, g, r, show=False)
    imgs_ncc.append(im_out)
    offsets_ncc.append(offset)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(20, 6))
for i in range(len(name_list)):
    axs[i].imshow(imgs_ssd[i])
    axs[i].axis('off')
    axs[i].set_title(name_list[i][:-4] + 
                     f"\n G displacement: {offsets_ssd[i][0]}\n R displacement: {offsets_ssd[i][1]}", 
                     fontsize=16, y=-0.2)

fig.suptitle("Single-scale Alignment with Euclidean Distance", fontsize=20)
fig.subplots_adjust(wspace=0)
plt.show()

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(20, 6))
for i in range(len(name_list)):
    axs[i].imshow(imgs_ncc[i])
    axs[i].axis('off')
    axs[i].set_title(name_list[i][:-4] + 
                     f"\n G displacement: {offsets_ncc[i][0]}\n R displacement: {offsets_ncc[i][1]}", 
                     fontsize=16, y=-0.2)

fig.suptitle("Single-scale Alignment with Normalized Cross-Correlation (NCC)", fontsize=20)
fig.subplots_adjust(wspace=0)
plt.show()
```

```python
def downsample(img):
    h, w = img.shape
    padded = np.pad(img, ((0, h % 2), (0, w % 2)), mode="reflect")
    h, w = padded.shape
    h_new, w_new = h // 2, w // 2
    out = np.zeros((h_new, w_new))
    out = (padded[0::2, 0::2] + padded[0::2, 1::2] + padded[1::2, 0::2] + padded[1::2, 1::2]) / 4.0
    return out

def gaussian_pyramid(b, g, r, kernel):
    #print('current shape ', b.shape)
    if b.shape[0] < 200 and b.shape[1] < 200: 
        # Reach the top level, do similarity search from the coarsest scale
        offset = single_ncc_offset(b, g, r, (-15, 16))
        return offset
    
    # Not at the top level, do downsample
    b_down = downsample(b)
    g_down = downsample(g)
    r_down = downsample(r)
    offset = gaussian_pyramid(b_down, g_down, r_down, kernel) # Recursion
    #print(offset)
    offset = single_ncc_offset(b, g, r, (-5, 6), offset=offset*2)
    #print('done shape ', b.shape)
    return offset

```

```python
def upsample(img):
    return np.repeat(np.repeat(img, 2, axis=0), axis=1)

def gaussian_pyramid_with_laplacian(b, g, r, kernel):
    #print('current shape ', b.shape)
    if b.shape[0] < 200 and b.shape[1] < 200: 
        # Reach the top level, do similarity search from the coarsest scale
        offset = single_ncc_offset(b, g, r, (-15, 16))
        return offset
    
    # Not at the top level, do downsample
    b_down = downsample(b)
    g_down = downsample(g)
    r_down = downsample(r)
    offset = gaussian_pyramid(b_down, g_down, r_down, kernel) # Recursion
    #print(offset)
    offset = single_ncc_offset(b, g, r, (-5, 6), offset=offset*2)
    #print('done shape ', b.shape)
    return offset
```

```python
def single_ncc_offset(b, g, r, window, offset=((0,0), (0,0))):
    #print(offset)
    max_l2_bg = -1
    max_l2_br = -1
    best_shift_bg = offset[0]
    best_shift_br = offset[1]
    for i in range(window[0], window[1]):
        for j in range(window[0], window[1]):
            off_g = (offset[0][0] + i, offset[0][1]+ j)
            shift_g = np.roll(g, off_g, (0, 1))
            l2_bg = ncc(b, shift_g)
            if l2_bg > max_l2_bg:
                max_l2_bg = l2_bg
                best_shift_bg = off_g

            off_r = (offset[1][0] + i, offset[1][1] + j)
            shift_r = np.roll(r, off_r, (0, 1))
            l2_br = ncc(b, shift_r)
            #print(j, l2_br)
            if l2_br > max_l2_br:
                max_l2_br = l2_br
                best_shift_br = off_r

    return np.array((best_shift_bg, best_shift_br))

def shift_with_offset_and_show(offset, b, g, r, show=True):
    ag = np.roll(g, offset[0], (0, 1))   
    ar = np.roll(r, offset[1], (0, 1))
    print(ag.shape, ar.shape, b.shape)
    im_out = np.dstack([ar, ag, b])
    if show:
        skio.imshow(im_out)
        plt.axis('off')
        plt.show()
    return im_out
```
